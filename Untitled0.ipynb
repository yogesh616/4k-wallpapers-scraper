```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create the data
# Generate some simple linear data (y = 2x - 1)
x_train = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
y_train = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

# Step 2: Build the model
# Create a simple sequential model with one dense layer
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])

# Step 3: Compile the model
# Compile the model using mean squared error as the loss function and stochastic gradient descent as the optimizer
model.compile(optimizer='sgd', loss='mean_squared_error')

# Step 4: Train the model
# Train the model for 500 epochs
model.fit(x_train, y_train, epochs=500)

# Step 5: Make predictions
# Use the model to predict a new value
x_test = np.array([10.0])
prediction = model.predict(x_test)

print(f"Prediction for x = 10.0: {prediction[0][0]}")

# Optional: Visualize the results
# Plot the training data and the model's prediction
plt.plot(x_train, y_train, 'ro', label='Training Data')
plt.plot(x_train, model.predict(x_train), label='Model Prediction')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
